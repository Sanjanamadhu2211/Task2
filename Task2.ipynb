```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

#load the dataset
df = pd.read_csv('/Titanic-Dataset.csv')

#basic info
print("Basic Info: \n", df.info())
print("Summary stats: \n",df.describe(include='all'))
print("Null values: \n",df.isnull().sum())

#histogram for numeric features
df.hist(figsize = (10,6) , bins = 20)
plt.suptitle("Histograms of Numeric Features")
plt.tight_layout()
plt.show()

#Boxplots for outlier detection
sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age')
plt.show()

sns.boxplot(x=df['Fare'])
plt.title('Boxplot of Fare')
plt.show()

#Correlation matrix
corr_matrix=df.corr(numeric_only=True)
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title("Correlation Matrix")
plt.show()

#Pairplot for feature relationships
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']].dropna(), hue='Survived')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()

#Count plot: Survival by Sex
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Count by Gender")
plt.show()

#Count plot: Survival by Pclass
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Count by Passenger Class")
plt.show()

#Interactive scatter plot
fig = px.scatter(df, x='Age', y='Fare', color='Survived', title="Age vs Fare (Survival)")
fig.show()
```
